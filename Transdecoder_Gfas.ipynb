```python
import gzip

def unzip_fasta_file(input_file, output_file):
    with gzip.open(input_file, 'rt') as gz_file:
        with open(output_file, 'w') as fasta_file:
            for line in gz_file:
                fasta_file.write(line)

# Usage example
input_file = 'gfas_final_1.0.fasta.gz'  # Path to the input gzipped FASTA file
output_file = 'gfas_final_1.0.fasta'  # Path to the output uncompressed FASTA file

unzip_fasta_file(input_file, output_file)
```
