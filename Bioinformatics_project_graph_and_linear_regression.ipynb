# Comparative Graph and Linear Regression 


```python
import matplotlib.pyplot as plt

species = ['Porities rus', 'Orbicella faveolata', 'Pocillopora damicornis', 'Galaxia fascicularis']
depths = [100, 20, 60, 80]
gfp_homologs = [5, 0, 9, 7]

bar_width = 0.35  # Width of the bars

depth_color = 'pink'
gfp_homologs_color = 'red'

plt.bar(species, depths, width=bar_width, label='Depth', color=depth_color)
plt.bar(species, gfp_homologs, width=bar_width, label='Number of GFP homologs', color=gfp_homologs_color)

plt.xlabel('Coral Species')
plt.ylabel('Depth / Number of GFP Homologs')
plt.title('Comparison of Depth and GFP Homologs')

plt.xticks(rotation=45)

plt.legend()

plt.tight_layout() 

# Add numbers to the bars
for i in range(len(species)):
    plt.text(species[i], depths[i], str(depths[i]), ha='center', va='bottom')
    plt.text(species[i], gfp_homologs[i], str(gfp_homologs[i]), ha='center', va='bottom')

plt.savefig('comparison_graph.png', transparent=True)

plt.show()
```
